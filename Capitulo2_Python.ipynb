## Operaciones con Arreglos y funciones

```python
# Establecer rangos en un Array
matriz = np.arange(1,50)
matriz
```

```python
matriz[:7]
```

```python
matriz[5:]
```

```python
matriz
```

```python
# max(): Obtener el valor mayor dentro del vector 
print('',matriz.max())
```

```python
# min(): Obtener el valor menor dentro del vector
print('',matriz.min())
```

```python
# round(): Redondea los valores del vector
matrizRound = np.linspace(0,50)
matrizRound
```

```python
# round(): Redondea los valores del vector
print('',matrizRound.round())
```

```python
# sum(): Nos devuelve la suma de todos los valores del vector
print('',matriz.sum())
```

```python
# put():Inserta un nuevo elemento en la matriz en la posición especificada
print(matriz)
```

```python
# Agrega una elemento a la Matriz
print('',matriz.put(-1, 50))
```

## condicionales

```python
n = 0  #  Asignación de 0 en n
while n < 10:  #  Expresión relacional n < 10, que devuelve True
    if (n % 2) == 0:  # Expresión aritmética y expresión relacional
        print(n,'es un número par') 
    else:
        print(n,'es un número impar')
    n += 1  # expresión aritmética n = n + 1 equivalente a operación en asignación n+=1
```
