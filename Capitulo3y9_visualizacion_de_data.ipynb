## Obtención de datos (Excel) y diferentes tipos de gráficas

```python
%pylab inline 
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('precision',5)
```

```python
#Cargando Datos de Excel
datosDemograficosExcel = pd.read_excel('Peliculas.xlsx', sheetname='Hoja3')
datosDemograficosExcel
```

```python
# Manejo de Columnas - 
print(datosDemograficosExcel['Cuidad'])
```

```python
# Manejo de Filas 
print(datosDemograficosExcel.ix[:])
```

** Ordenamiento de Datos - Sort_values **

```python
# Ordenando los Datos segun la Mortalidad
print(datosDemograficosExcel.sort_values(by='Año'))
```

```python
# Trabajando solo con una Columna
EsperanzaVida = datosDemograficosExcel['Facebook Likes - Actor 1']
```

```python
EsperanzaVida
```

```python
# APlicando solo a una columna
EsperanzaVida=EsperanzaVida[EsperanzaVida<20]
EsperanzaVida
```

```python
EsperanzaVida.describe()
```

** Graficando los Resultados **

```python
# 
EsperanzaVida.plot()
```

```python
# Grafico de Barras
EsperanzaVida.plot(kind='bar')
```

```python
# Grafico de Densidad
EsperanzaVida.plot(kind='kde')
```

```python
# Grafico de Area
EsperanzaVida.plot(kind='area')
```

```python
# Grafico de Pie
EsperanzaVida.plot(kind='pie')
```

```python
# Graficando el Histograma
EsperanzaVida.plot(kind='hist')
```

```python
# Graficando varias Columnas
Grafico = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])
Grafico.plot.bar();
```

** Histogramas **

```python
# Creando un Histograma con 3 series
Histograma = pd.DataFrame({'a': np.random.randn(1000) + 1, 
                    'b': np.random.randn(1000),
                    'c': np.random.randn(1000) - 1}, 
                   columns=['a', 'b', 'c'])
plt.figure();
```

```python
Histograma.plot.hist(alpha=0.7)
```

```python
# Creando 500 Series diferentes
data = pd.Series(np.random.randn(500))
#data
```

```python
# Graficando un Rango de Series en Histogramas
data.hist(by=np.random.randint(0, 6, 500), figsize = (10, 10))
```

** Box Plost - Graficos de Cajas **

```python
# Data Random  
graficoBox = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
graficoBox.plot.box()
```

```python
# Plot Box en vertical
graficoBox.plot.box(vert=False, positions=[1, 2, 3, 4, 7])
```

** graficando Series **

```python
# En Pandas un Dataframe son unidades basicas, una seria de datos
serieDatos = pd.Series(np.random.randn(7),name='NumeroRandom')
```

```python
# Imprimiendo y formateando a 2 decimales, creación automatica de indice
#print(serieDatos.round(2))
print(serieDatos)
```

```python
# Describe el DataFrame
print('Describir DataFrame', serieDatos.describe())
```

```python
# Graficando los Resultados
serieDatos.plot()
```

** Graficando Data Frame **

```python
serieDatos02 = pd.Series(np.random.randn(7),name='NumeroRandom')
#Sumatoria de Series
serieResultado =  serieDatos + serieDatos02
#serieResultado =  serieDatos + serieDatos02 * 7
print(serieResultado)
```

```python
# Creando Nuestro Primer DataFrame
series = {
    "Col1": serieDatos,
    "Col2": serieDatos02,
    "Col3": serieResultado
}
```

```python
primerDataFrame = pd.DataFrame(series)
print(primerDataFrame)
```

```python
primerDataFrame.plot()
```
