# Ficheros de texto
## Crear fichero y escribir texto

```python
# fichero.txt ruta donde lo crearemos, w indica modo de escritura, write (puntero principio)
fichero = open('Texto01.txt','w')  
```

```python
contenido = "Mi primer archivo de Texto  \nPython en DMC"
```

```python
# escribimos el texto
fichero.write(contenido) 
```

```python
# cerramos el fichero
fichero.close()  
```

## Lectura línea a línea secuencial

```python
with open("Texto01.txt", "r") as fichero:
    for linea in fichero:
        print(linea)
```

## Manejando el puntero

```python
fichero = open('Texto01.txt','r')
fichero.seek(0) # Puntero al principio
fichero.read(7) # Leemos 10 carácteres
```

```python
fichero.read(15) # Leemos 10 carácteres más, a partir del 10 donde está el puntero
```

```python
fichero.seek(0)
fichero.seek( len(fichero.readline()) ) # Leemos la primera línea y situamos el puntero al principio de la segunda
```

```python
fichero.read() # Leemos todo lo que queda del puntero hasta el final
```

## Lectura y escritura a la vez

```python
fichero2 = open('fichero2.txt','w')
```

```python
texto = "Panama 1\nColombia 2\nBrasil 3\nPuerto rico 4"
```

```python
fichero2.write(texto)
```

```python
fichero2.close()
```

```python
fichero2 = open('fichero2.txt','r+')  # + escritura simultánea, puntero al principio por defecto
```

```python
fichero2.write('Mexico 5')
```

```python
fichero2.close()
```

## Modificar una línea específica

```python
# modo lectura con escritura, puntero al principio por defecto
fichero2 = open('fichero2.txt','r+')  
```

```python
# leemos todas las líneas
texto = fichero2.readlines() 
```

```python
texto[2] = "Francia\n"  # indice menos 1
```

```python
texto
```

```python
fichero2.seek(0) # Ponemos el puntero al principio
fichero2.writelines(texto)
```

```python
fichero2.close()
```
